```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
Unemployment = [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
plt.plot(Year, Unemployment)
plt.title('Unemployment Rate Vs Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.scatter(Year,Unemployment)
plt.title('Unemployment Rate Vs Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.bar(Year,Unemployment,color="green",width=5)
plt.title('Unemployment Rate Vs Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
sns.set(style="darkgrid")
df = sns.load_dataset("iris")
sns.histplot(data=df, x="sepal_length", color="skyblue", label="Sepal Length", kde=True)
plt.legend() 
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
ax = sns.countplot(x="class", hue="who", data=titanic)
```

```python
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
```

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```

```python
df = sns.load_dataset('tips')
sns.pairplot(df, hue ='day')
plt.show()
```

```python
sns.jointplot(x = Year, y = Unemployment,
              kind = "kde")
```
